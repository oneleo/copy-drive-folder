```
# Mount drive to colab 掛載雲端硬碟
from google.colab import drive
import os
drive.mount('/content/gdrive')
# 假設要將檔案複製到雲端硬碟的 Colab Notebooks/_Copy_Folder 資料夾
!mkdir -p "/content/gdrive/My Drive/Colab Notebooks/_Copy_Drive_Folder/"
!chmod 755 "/content/gdrive/My Drive/Colab Notebooks/_Copy_Drive_Folder"
!ls -al "/content/gdrive/My Drive/Colab Notebooks/_Copy_Drive_Folder"
```

```
# [Important] Paste the Shared Folder shortcut name below【重要】請將放在 Colab Notebooks/_Copy_Folder 內的捷徑完整名稱填在下方雙引號內
os.environ['TARGET'] = "Shared Folder Sample"
!echo ${TARGET}
```

```
# Copy drive folder 複製目標資料夾
!mkdir -p "/content/gdrive/My Drive/Colab Notebooks/_Copy_Drive_Folder/${TARGET}_copy"
!cp -r "$(readlink -f "/content/gdrive/My Drive/Colab Notebooks/_Copy_Drive_Folder/${TARGET}")" "/content/gdrive/My Drive/Colab Notebooks/_Copy_Drive_Folder/${TARGET}_copy"
```

```
# Compress drive folder 壓縮目標資料夾
!zip -rq9 "/content/gdrive/My Drive/Colab Notebooks/_Copy_Drive_Folder/${TARGET}.zip" "$(readlink -f "/content/gdrive/My Drive/Colab Notebooks/_Copy_Drive_Folder/${TARGET}")"
```
